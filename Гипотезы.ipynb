__Гипотеза 1. Котировки доллара__

H1: в дни налогового периода (20-28 число каждого месяца) доллар слабеет по отношению к рублю (эскпортеры продают валютную выручку)
Н0: в дни налогового периода котировки валюты не реагируют на продажу выручки

Ошибка 2 рода: Пропускаем рост рубля и возможность для шорта/фиксации прибыли на пиках. -> возможный убыток по спекуляциям
Ошибка 1 рода: Шортим/фиксируем, котировки во флэте или идут вверх. -> возможную прибыль уже получили или излишне перестраховались, закрыв убыточную позицию, но сверх этого ничего не потеряв. (безопаснее, чем уйти в минус, если рубль начнет резко слабеть)



__Гипотеза 2. Котировки акции__

Н0: котировки Сургутнефтегаз меняются безотносительно цен на нефть, газ и валюту
Н1: котировки Сургутнефтегаз растут положительно/отрицательно вместе с ростом цены на фьючерс на нефть, газ, а также вместе с котировкой доллара на фрейме 1 рабочей недели* (потому что у компании есть огромные накопления в долларе, а также она занимается поставками нефти и газа)

Ошибка 1 рода: отвергаем нулевую гипотезу, а она оказывается верна. Зная прогнозы мировых инвестбанков, инвестируем в Сургут, а он, по воле небесной, начинает падать. -> теряем деньги, становимся долгосрочным инвестором в бесперспективный актив.
Ошибка 2 рода: не отвергаем нулевую гипотезу, а она неверна. Упускаем возможность заработать на инвестиции, которая считается по логичным прогнозам других признаков, но, по крайней мере, ничего не теряем (не так страшно, как в ошибке 1 рода)



__Гипотеза 3. Сбербанк__

Н0: Сургутнефтегаз растет, когда растет Сбербанк 
Н1: Сургутнефтегаз живет своей жизнью, не реагируя на котировки Сбера - самой популярной среди домашних инвесторов-кухарок акции или индикатора здоровья и прокси всего российского фондового рынка

Ошибка 1 рода: не вкладываемся в Сургут, не веря в светлое будущее нефтегаза даже при растущей уверенности в российском бизнесе. -> теряем инвествозможность, страдаем
Ошибка 2 рода: видим рост Сбера и надеемся получить гейнс на компании ВВП, а инвесторы решают его подслить и переложить деньги во что-то более интересное. -> теряем деньги и страдаем еще больше от того, что вложились в неликвидную и никому не нужную фигню((
