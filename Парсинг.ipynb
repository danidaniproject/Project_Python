### SNGS

```python
#вставляем ссылку с нужным инструментом и датой
url = 'https://www.moex.com/ru/marketdata/#/mode=instrument&secid=SNGS&boardgroupid=57&mode_type=history&date_from=2022-10-01&date_till=2023-05-01'
```

```python
#жмём, откроется гугл бразуер не пугаемся этого и не закрываем его
driver = webdriver.Chrome('chromedriver')
driver.implicitly_wait(30)
```

```python
#жмём
driver.get(url)
#после открытия страницы нажимаем в бразуере "Согласен"
```

```python
df = pd.DataFrame()
```

```python
#читаем первую страницу
df = pd.read_html(driver.find_element(By.CLASS_NAME, "ui-table").get_attribute('outerHTML'))[0]
```

```python
#пишем сколько страниц всего
i = 102
```

```python
c = 0
while c < i-1: 
    print(c)
    driver.find_element(By.XPATH, '//*[@id="instrument-table"]/div[1]/div[2]/div/div/div/button[9]/span').click()
    time.sleep(3)
    df1 = pd.read_html(driver.find_element(By.CLASS_NAME, "ui-table").get_attribute('outerHTML'))[0]
    df = df.append(df1)
    c += 1
```

```python
df #не прочитались запятые почему-то, ниже исправим проблему
```

```python
def tryconvert(x): #чтобы пропускать ошибки
    try:
        y = float(str(x)[:2] + '.' + str(x)[2:])
        return y
    except:
        pass
```

```python
df['Срвзв.цена'] = df['Срвзв.цена'].apply(lambda x: tryconvert(x))
```

```python
df['Первая'] = df['Первая'].apply(lambda x: tryconvert(x))
```

```python
df['Минимум'] = df['Минимум'].apply(lambda x: tryconvert(x))
```

```python
df['Максимум'] = df['Максимум'].apply(lambda x: tryconvert(x))
```

```python
df['Закрытия'] = df['Закрытия'].apply(lambda x: tryconvert(x))
```

```python
df['Объем'] = df['Объем'].apply(lambda x: x.replace('\xa0', "").replace(',', '.'))
```

```python
def vol(x):
    try:
        return(float(x))
    except:
        pass
```

```python
df['Объем'] = df['Объем'].apply(lambda x: vol(x))
```

```python
df = df[['Дата торгов', 'Код инструмента', 'Сделок, шт.', 'Объем', 'Срвзв.цена']]
```

```python
df.to_csv('SNGS_2015_2023.csv')
```
