```python
import numpy as np
import pandas as pd

import scipy.stats 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sgns = pd.read_csv('SNGS_okt-may_v2.csv')
sgns
```

```python
si = pd.read_csv('Si_okt-may_v2.csv')
si
```

```python
dates_taxes = sgns['Дата торгов'].to_frame()

dates_taxes['День'] = dates_taxes['Дата торгов'].apply(lambda x: int(x.replace('-', ""))%100)
dates_taxes #нашли день месяца, чтобы далее понять, относится ли он к налоговому периоду
```

```python
#примем за условие, что валютная выручка в среднем продается 22-28 числа, и в этот момент рубль укрепляется

dates_taxes['Налоговый период'] = np.where(((dates_taxes['День']>=22) & (dates_taxes['День']<=28)), 1, 0)
siprice = si[['Средне-взвешеннаяцена за лот', 'Дата']]
dt_tx_pr = dates_taxes.merge(sgprice, how='left', on='Дата торгов')
sip = siprice['Средне-взвешеннаяцена за лот']
dates_taxes
```

```python
df_r = sip.diff() / sip.shift(1)
df_r = df_r.iloc[1:]
df_r = df_r.to_frame()

df_r['Дата торгов'] = dates_taxes['Дата торгов'][1:]
df_r['Динамика'] = np.where(df_r['Средне-взвешеннаяцена за лот']>=0, 0, 1)
df_r = df_r.rename(columns={"Средне-взвешеннаяцена за лот": "Процентное изменение"})
aggreg = df_r.merge(dates_taxes, on='Дата торгов', how='left')
hypotesis1 = aggreg[['Динамика', 'Налоговый период']]
hypotesis1 #вытащили признаки, которые нам нужны для конфьюжн матрикс, учитывая, что 1 и 1 - это условие роста рубля
```

Для нашей гипотезы предположим, что если рубль укрепляется, то это 1. Если налоговый период, то также 1. То есть при 1 и 1 или 0 и 0 наша гипотеза верна. Построим таблицу.

```python
ct_o = hypotesis1.groupby('Налоговый период')['Динамика'].value_counts().unstack()
ct_o
```

```python
accuracy = (72+13)/(72+13+39+21)

print(f'Accuracy = {accuracy}') #рандомайзер получается с нашими правильными ответами
```

```python
scipy.stats.chi2_contingency(ct_o, correction=False)
```

Видим, что результаты очень плохие ;) Это потому что у нас мало наблюдений, а еще потому что в нашей гипотезе неверно обратное: если налогового периода нет, то вообще-то это совсем не значит, что рубль будет слабеть. А вот у обратного есть теоретическое подкрепление, но для российского рынка оно не работает. Всегда есть факторы того, что макроэкономика неактуальна - очень много разных политических событий, а также штук, подконтрольных ЦБ, влияют на курс рубля и доллара. Однако если смотреть на горизонте пошире, ошибка будет уменьшаться.
